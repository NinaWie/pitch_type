# Evaluation of parameters

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import scipy as sp
import scipy.stats

#from sklearn.preprocessing import StandardScaler
from data_preprocess import Preprocessor
from tools import Tools
from model import Model
import matplotlib.pylab as plt

import shutil
```

```python
df = pd.read_csv("test_parameters.csv")
columns = df.columns.tolist()
print(columns)
```

```python
#df = df.drop(df.index[[0,1,2,3,4,5]])
acc = df["highest_acc"].values
print(acc)
columns = ['Batchsize', 'Epochs', 'align', 'dropout', 'file', 'learning_rate', 'len_train', 
           'lstm_hidden_layers', 'lstm_units', 'model', 'nr_classes', 'act']

plt.rcParams.update({'font.size': 18})
for col in columns:
    act = np.unique(df[col].values)
    cut = [np.mean((df[df[col]==elem])["highest_acc"].values) for elem in act]

    plt.figure(figsize=(20,10))
    plt.ylim([0.4,0.7])
    y_pos = range(len(act))
    plt.bar(y_pos, cut, align='center', alpha=0.5, width = 0.5)
    #plt.ylim([0,350])
    plt.xticks(y_pos, act)
    plt.ylabel('highest test accuracy achieved')
    plt.title(col)
    plt.show()
```
